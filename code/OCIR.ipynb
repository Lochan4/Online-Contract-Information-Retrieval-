```python
import requests
from bs4 import BeautifulSoup
import csv

##Initating values to be none
estimate = None
city=None
contract_id=None
description = None
business_type = None
license_type = None

URL = "https://ppmoe.dot.ca.gov/des/oe/weekly-ads/all-adv-projects.php"
HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})
webpage = requests.get(URL, headers=HEADERS)
soup = BeautifulSoup(webpage.content, "html.parser")

# Get the projects div
projects_div = soup.find("div", attrs={"class": "projects"})
project_containers = projects_div.find_all("div", attrs={"class": "project-container"})

# CSV Setup
csv_filename = "project_data.csv"
fieldnames = ['Contract ID', 'Description', 'Estimate', 'City', 'License Type', 'Business Type']
with open(csv_filename, 'w', newline='') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()

    for container in project_containers:
        container_text = container.text
        
        lines = container_text.split("\n")
        contract_id = lines[0].split("|")[0].strip()
        description = lines[1].strip()
        
        # Extracting Estimate and City (Corrected)
        for line in lines:
            if "Estimate:" in line:
                estimate = line.split("Estimate: $")[1].strip()[:8]
                # estimator = r"\d{3,}"
                # estimate = re.findall(estimator, line)
                city = line[line.find("In") + 3 : line.find("In") + 13].strip()
                break
        
        # Improved License Type Extraction
        license_type = "N/A"  # Default value
        for line in lines:
            if "license" in line.lower():  # Case-insensitive search
                if "class a" in line.lower():
                    license_type = "Class A"
                elif "class c" in line.lower():
                    license_type = line[line.lower().find("class c"):].strip() # Get the exact text after "class c"
                elif "class b" in line.lower():
                    license_type = "Class B"
                # Add more conditions for other license types if needed
                break
        # Extracting Business Type
        business_type = "N/A"  # Default value
        for line in lines:
            if "small business preference" in line.lower():
                business_type = "Small Business Preference"
                break

        # Write data to CSV
        writer.writerow({
            'Contract ID': contract_id,
            'Description': description,
            'Estimate': estimate,
            'City': city,
            'License Type': license_type,
            'Business Type': business_type
        })

print(f"CSV file '{csv_filename}' created successfully.")
```
